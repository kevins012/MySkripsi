```python
import tensorflow as tf
import numpy as np

# Dataset: 4 gambar 64x64x3 acak dan 4 label
images = np.random.rand(4, 64, 64, 3)  # 4 gambar berukuran 64x64x3
labels = np.array([0, 1, 2, 3])  # 4 label acak (untuk klasifikasi 4 kelas)

# Mengonversi menjadi tf.data.Dataset
dataset = tf.data.Dataset.from_tensor_slices((images, labels))

# Menampilkan elemen pertama dari dataset
for image, label in dataset.take(1):
    print("Image shape:", image.shape)  # Menampilkan shape gambar
    print("Label:", label)  # Menampilkan label
for image, label in dataset.take(1):
    print("Image as array:", image.numpy())  # Menampilkan gambar sebagai array
    print("Label as array:", label.numpy())  # Menampilkan label sebagai array
```

```python
import tensorflow as tf

# Fungsi generator yang menghasilkan data
def generate_data():
    for i in range(5):
        yield (i, i * 2)  # Menghasilkan pasangan (x, y)

# Mengonversi generator ke tf.data.Dataset dengan output_signature yang benar
dataset = tf.data.Dataset.from_generator(
    generate_data, 
    output_signature=(tf.TensorSpec(shape=(), dtype=tf.int32), tf.TensorSpec(shape=(), dtype=tf.int32))
)

# Menampilkan elemen pertama dari dataset
for x, y in dataset.take(1):
    print("x:", x.numpy(), "y:", y.numpy())
```

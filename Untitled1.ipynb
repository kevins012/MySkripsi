```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan vektor
A = np.array([0, 0])  # Titik tetap
C = np.array([5, 3])  # Titik yang akan dirotasikan
B = np.array([3, 1])
D = np.array([-3, 7])

# Hitung sudut antara AC dan sumbu X
AC = C - A
theta = np.arctan2(AC[1], AC[0])  # Sudut terhadap sumbu X
theta_rotasi = np.pi / 2 - theta  # Agar tegak lurus terhadap X

# Matriks rotasi 2D
R = np.array([
    [np.cos(theta_rotasi), -np.sin(theta_rotasi)],
    [np.sin(theta_rotasi),  np.cos(theta_rotasi)]
])

# Rotasi semua titik
C_rotated = np.dot(R, C)
B_rotated = np.dot(R, B)
D_rotated = np.dot(R, D)

# Fungsi untuk menggambar vektor
def plot_vector(ax, start, end, color, label, linestyle='-', linewidth=2):
    ax.arrow(
        start[0], start[1], end[0] - start[0], end[1] - start[1],
        head_width=0.1, head_length=0.2, fc=color, ec=color,
        linestyle=linestyle, linewidth=linewidth, label=label, alpha=0.8
    )

# Visualisasi
fig, ax = plt.subplots(figsize=(8, 8))

# Plotkan vektor asli (garis putus-putus, lebih tipis)
plot_vector(ax, A, B, 'blue', 'AB (Original)', linestyle='--', linewidth=1)
plot_vector(ax, A, C, 'red', 'AC (Original)', linestyle='--', linewidth=1)
plot_vector(ax, A, D, 'green', 'AD (Original)', linestyle='--', linewidth=1)

# Plotkan vektor setelah rotasi (garis solid, lebih tebal)
plot_vector(ax, A, B_rotated, 'cyan', 'AB (Rotated)', linestyle='-', linewidth=2)
plot_vector(ax, A, C_rotated, 'orange', 'AC (Rotated)', linestyle='-', linewidth=2)
plot_vector(ax, A, D_rotated, 'lime', 'AD (Rotated)', linestyle='-', linewidth=2)

# Pengaturan tampilan
ax.set_xlim([-12, 12])
ax.set_ylim([-12, 12])
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_aspect('equal', 'box')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Rotasi Vektor terhadap AC')
ax.legend()

# Tampilkan grid dan plot
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Fungsi untuk menghitung panjang vektor
def vector_length(vector):
    return np.linalg.norm(vector)

# Panjang vektor sebelum rotasi
length_B_original = vector_length(B)
length_C_original = vector_length(C)
length_D_original = vector_length(D)

# Panjang vektor setelah rotasi
length_B_rotated = vector_length(B_rotated)
length_C_rotated = vector_length(C_rotated)
length_D_rotated = vector_length(D_rotated)

# Cetak panjang vektor sebelum dan sesudah rotasi
print("Panjang vektor sebelum rotasi:")
print(f"AB: {length_B_original:.2f}, AC: {length_C_original:.2f}, AD: {length_D_original:.2f}")

print("\nPanjang vektor setelah rotasi:")
print(f"AB (Rotated): {length_B_rotated:.2f}, AC (Rotated): {length_C_rotated:.2f}, AD (Rotated): {length_D_rotated:.2f}")

# Verifikasi bahwa panjangnya tetap sama
print("\nVerifikasi bahwa panjang vektor tetap:")
print(f"AB: {'Sama' if np.isclose(length_B_original, length_B_rotated) else 'Berbeda'}")
print(f"AC: {'Sama' if np.isclose(length_C_original, length_C_rotated) else 'Berbeda'}")
print(f"AD: {'Sama' if np.isclose(length_D_original, length_D_rotated) else 'Berbeda'}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik awal
points = np.array([
    [1, 2],  # A
    [5, 2],  # B
    [9, 6],  # C
    [3, 2],  # D
    [2, 1],
    [4,7],
    [2,5]# E
])

# Tentukan bounding box
x_min, y_min = np.min(points, axis=0)  # Titik kiri atas bounding box (pivot)
x_max, y_max = np.max(points, axis=0)  # Titik kanan bawah bounding box

# Dimensi bounding box
width = x_max - x_min
height = y_max - y_min

# Titik target untuk skala (x_target, y_target)
x_target = 12
y_target = 8
scale_x = (x_target - x_min) / width
scale_y = (y_target - y_min) / height
scale = min(scale_x, scale_y)  # Jaga proporsi dengan menggunakan skala terkecil

# Transformasi titik
transformed_points = points.copy()
for i, point in enumerate(points):
    transformed_points[i] = [
        x_min + scale * (point[0] - x_min),
        y_min + scale * (point[1] - y_min)
    ]

# Visualisasi
plt.figure(figsize=(10, 8))

# Bounding box awal
plt.plot([x_min, x_max, x_max, x_min, x_min],
         [y_min, y_min, y_max, y_max, y_min],
         color='blue', linestyle='--', label='Bounding Box Awal')

# Bounding box baru
new_x_max = x_min + scale * width
new_y_max = y_min + scale * height
plt.plot([x_min, new_x_max, new_x_max, x_min, x_min],
         [y_min, y_min, new_y_max, new_y_max, y_min],
         color='red', linestyle='--', label='Bounding Box Baru')

# Plot titik awal dan garis vektor
for i, point in enumerate(points):
    # Titik awal
    plt.scatter(point[0], point[1], color='blue', label="Titik Awal" if i == 0 else "")
    plt.text(point[0], point[1], f"{chr(65 + i)}", fontsize=10, color="blue")
    # Garis vektor dari pivot ke titik awal
    plt.plot([x_min, point[0]], [y_min, point[1]], color='blue', linestyle='-', alpha=0.7, label="Vektor Awal" if i == 0 else "")

# Plot titik baru dan garis vektor
for i, point in enumerate(transformed_points):
    # Titik setelah transformasi
    plt.scatter(point[0], point[1], color='red', label="Titik Baru" if i == 0 else "")
    plt.text(point[0], point[1], f"{chr(65 + i)}'", fontsize=10, color="red")
    # Garis vektor dari pivot ke titik transformasi
    plt.plot([x_min, point[0]], [y_min, point[1]], color='red', linestyle='--', alpha=0.7, label="Vektor Baru" if i == 0 else "")

# Garis bantu vertikal dan horizontal
plt.axvline(x=x_min, color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=y_min, color='gray', linestyle='--', alpha=0.5)

# Pengaturan plot
plt.title("Transformasi Skala dengan Vektor dan Bounding Box")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```

```python
def rotate_point(x, y, a, b, theta_deg):
    """
    Rotasi titik (x, y) terhadap pusat (a, b) sebesar theta_deg derajat (berlawanan arah jarum jam).
    
    Parameters:
    - x, y: koordinat titik awal
    - a, b: koordinat pusat rotasi
    - theta_deg: besar sudut rotasi dalam derajat
    
    Returns:
    - (x', y'): koordinat titik setelah rotasi
    """
    # Konversi derajat ke radian
    theta_rad = np.deg2rad(theta_deg)
    
    # Translasi titik ke origin
    x_shifted = x - a
    y_shifted = y - b
    
    # Rotasi
    x_rotated = x_shifted * np.cos(theta_rad) - y_shifted * np.sin(theta_rad)
    y_rotated = x_shifted * np.sin(theta_rad) + y_shifted * np.cos(theta_rad)
    
    # Translasi balik
    x_new = x_rotated + a
    y_new = y_rotated + b
    
    return x_new, y_new
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Garis awal (koordinat x1, y1, x2, y2)
lines = np.array([
    [2, 2, 6, 2],  # Garis horizontal
    [6, 2, 6, 6],  # Garis vertikal
    [6, 6, 2, 6],  # Garis horizontal atas
    [2, 6, 2, 2],  # Garis vertikal kiri
    [3, 3, 5, 5],  # Garis diagonal
])

# Pusat transformasi (titik tetap)
pivot = np.array([0, 0])

# Skala yang diterapkan
scale = 2

# Fungsi untuk menskalakan garis
def scale_lines(lines, pivot, scale):
    scaled_lines = []
    for line in lines:
        p1 = line[:2] - pivot  # Titik awal relatif terhadap pivot
        p2 = line[2:] - pivot  # Titik akhir relatif terhadap pivot
        p1_scaled = p1 * scale + pivot  # Skalakan dan kembalikan ke posisi asli
        p2_scaled = p2 * scale + pivot
        scaled_lines.append(np.concatenate((p1_scaled, p2_scaled)))
    return np.array(scaled_lines)

# Garis setelah skala
scaled_lines = scale_lines(lines, pivot, scale)

# Visualisasi
fig, ax = plt.subplots(figsize=(8, 8))

# Garis awal (sebelum skala)
for line in lines:
    ax.plot([line[0], line[2]], [line[1], line[3]], color='blue', label='Awal', alpha=0.7)
    
# Garis setelah skala
for line in scaled_lines:
    ax.plot([line[0], line[2]], [line[1], line[3]], color='red', linestyle='--', label='Setelah Skala', alpha=0.7)

# Visualisasi titik pivot
ax.scatter(pivot[0], pivot[1], color='black', marker='o', label='Pivot', zorder=5)

# Konfigurasi plot
ax.axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax.axvline(0, color='gray', linewidth=0.5, linestyle='--')
ax.set_aspect('equal', adjustable='datalim')
ax.legend(['Awal', 'Setelah Skala'], loc='upper left')
ax.set_title('Transformasi Skala Seperti MS Word')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

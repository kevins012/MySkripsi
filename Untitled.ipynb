```python
import pandas as pd
import numpy as np

# ====== 1. Buat data landmark manual (3 baris, 21 titik, masing-masing X,Y,Z) ======
# data_manual = {
#     **{f'X{i}': [0.5 + i * 0.01 + j * 0.02 for j in range(3)] for i in range(21)},
#     **{f'Y{i}': [0.5 + i * 0.015 - j * 0.01 for j in range(3)] for i in range(21)},
#     **{f'Z{i}': [0.5 + i * 0.005 + j * 0.015 for j in range(3)] for i in range(21)},
# }
# df_manual = pd.DataFrame(data_manual)

# ====== 2. Fungsi preprocessing berdasarkan metode transformasi basis lokal ======
def preprocess_hand_landmarks(df):
    n_rows = df.shape[0]
    processed = []

    for idx in range(n_rows):
        # Ambil titik-titik (21x3)
        points = np.array([[df.loc[idx, f'X{i}'], df.loc[idx, f'Y{i}'], df.loc[idx, f'Z{i}']] for i in range(21)])

        # Translasi ke titik 0 sebagai origin
        origin = points[0]
        translated = points - origin

        # Buat basis lokal dari titik 5 dan 17
        v1 = translated[5]
        v2 = translated[17]

        x_axis = v1 / np.linalg.norm(v1)
        y_temp = v2 / np.linalg.norm(v2)
        z_axis = np.cross(x_axis, y_temp)
        z_axis /= np.linalg.norm(z_axis)
        y_axis = np.cross(z_axis, x_axis)
        y_axis /= np.linalg.norm(y_axis)

        # Matriks basis lokal (kolom = basis vector)
        R = np.vstack([x_axis, y_axis, z_axis]).T

        # Proyeksikan semua titik ke basis lokal
        projected = translated @ R  # shape (21, 3)

        # Hitung panjang ruas (bone length) untuk beberapa segmen penting
        bone_pairs = [(0, 5), (0, 17), (5, 6), (6, 7), (7, 8), (17, 18), (18, 19), (19, 20), (0, 9), (9, 10)]
        bone_lengths = [np.linalg.norm(translated[i] - translated[j]) for i, j in bone_pairs]

        # Hitung beberapa sudut antara segmen jari
        def angle_between(a, b):
            cos_theta = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
            return np.arccos(np.clip(cos_theta, -1.0, 1.0))  # radian

        angles = [
            angle_between(translated[5] - translated[0], translated[17] - translated[0]),
            angle_between(translated[5] - translated[6], translated[6] - translated[7]),
            angle_between(translated[17] - translated[18], translated[18] - translated[19])
        ]

        # Gabungkan fitur
        features = {
            **{f'Px{i}': projected[i][0] for i in range(21)},
            **{f'Py{i}': projected[i][1] for i in range(21)},
            **{f'Pz{i}': projected[i][2] for i in range(21)},
            **{f'bone_len_{i}': bl for i, bl in enumerate(bone_lengths)},
            **{f'angle_{i}': ang for i, ang in enumerate(angles)},
            'Label': df.loc[idx, 'Label']  # Tambahkan label asli
        }
        processed.append(features)

    return pd.DataFrame(processed)

# ====== 3. Jalankan preprocessing ======
processed_df_manual = preprocess_hand_landmarks(df)
```

```python
def extract_features_from_landmarks(hand_landmarks):
    # Ambil koordinat x, y, z dari setiap titik landmark
    points = np.array([[lm.x, lm.y, lm.z] for lm in hand_landmarks])

    # Translasi titik 0 ke origin (posisi wrist jadi (0,0,0))
    origin = points[0]
    translated = points - origin

    # Basis lokal dari titik 5 (telunjuk) dan 17 (kelingking)
    v1 = translated[5]  # Titik ke-5 (telunjuk)
    v2 = translated[17]  # Titik ke-17 (kelingking)

    # Menormalisasi vektor
    x_axis = v1 / np.linalg.norm(v1)
    y_temp = v2 / np.linalg.norm(v2)
    z_axis = np.cross(x_axis, y_temp)
    z_axis /= np.linalg.norm(z_axis)
    y_axis = np.cross(z_axis, x_axis)
    y_axis /= np.linalg.norm(y_axis)

    # Matriks rotasi untuk sistem koordinat lokal
    R = np.vstack([x_axis, y_axis, z_axis]).T
    projected = translated @ R  # Proyeksi ke sistem koordinat lokal

    # Fitur yang dipilih sesuai dengan urutan yang diberikan
    features = []

    # Ambil koordinat Px, Py dari titik yang terpilih
    feature_order = [
        (4, 'Px'), (4, 'Py'), (8, 'Px'), (8, 'Py'), (12, 'Px'), (12, 'Py'),
        (16, 'Px'), (16, 'Py'), (20, 'Px'), (20, 'Py'),
        (5, 'Px'), (5, 'Py'), (9, 'Px'), (9, 'Py'), (13, 'Px'), (13, 'Py'),
        (17, 'Px'), (17, 'Py'), (1, 'Px'), (1, 'Py'), (2, 'Px'), (2, 'Py'),
        (3, 'Px'), (3, 'Py'), (10, 'Px'), (10, 'Py'), (14, 'Px'), (14, 'Py'),
        (18, 'Px'), (18, 'Py'),
        (6, 'Px'), (6, 'Py'), (7, 'Px'), (7, 'Py'), (11, 'Px'), (11, 'Py'),
        (15, 'Px'), (15, 'Py'), (19, 'Px'), (19, 'Py')
    ]
    
    for idx, coord in feature_order:
        features.append(projected[idx][0] if coord == 'Px' else projected[idx][1])  # Ambil Px atau Py

    # Panjang ruas (bone lengths) yang terpilih
    bone_pairs = [(0, 5), (0, 17), (5, 6), (6, 7), (7, 8), (17, 18), (18, 19), (19, 20), (0, 9), (9, 10)]
    bone_lengths = [np.linalg.norm(translated[i] - translated[j]) for i, j in bone_pairs]
    
    # Menambahkan panjang ruas
    features.extend(bone_lengths)

    # Sudut antar ruas jari
    def angle_between(a, b):
        cos_theta = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
        return np.arccos(np.clip(cos_theta, -1.0, 1.0))

    angles = [
        angle_between(translated[5] - translated[0], translated[17] - translated[0]),
        angle_between(translated[5] - translated[6], translated[6] - translated[7]),
        angle_between(translated[17] - translated[18], translated[18] - translated[19])
    ]

    # Menambahkan sudut
    features.extend(angles)

    # Debug: Periksa dimensi fitur sebelum dikirim ke model
    print(f"Fitur yang diekstrak: {len(features)}")
    return np.array([features], dtype=np.float32)  # Bentuk fitur akhir (1, 40)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, Flatten, Dense, LSTM
from tensorflow.keras.models import Model

# 1. Persiapan data (dummy data dengan jumlah sampel berbeda)

# Simbol Input (gambar simbol)
X_symbol = np.random.rand(100, 10, 10, 3)  # 100 gambar simbol acak
y_symbol = np.random.randint(0, 3, size=(100, 1))  # 0: ikan, 1: monyet, 2: burung

# Gerakan Input (urutan data sekuensial)
X_movement = np.random.rand(80, 5, 3)  # 80 urutan gerakan (5 frame, 3 fitur per frame)
y_movement = np.random.randint(0, 2, size=(80, 1))  # 0: gajah, 1: kucing

# 2. Model untuk simbol (statis)
input_symbol = Input(shape=(10, 10, 3))  # Ukuran gambar 10x10, 3 saluran (RGB)
x_symbol = Conv2D(32, (3, 3), activation='relu')(input_symbol)
x_symbol = Flatten()(x_symbol)
x_symbol = Dense(64, activation='relu')(x_symbol)
output_symbol = Dense(3, activation='softmax', name='symbol_output')(x_symbol)  # 3 kelas untuk simbol

model_symbol = Model(inputs=input_symbol, outputs=output_symbol)
model_symbol.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 3. Model untuk gerakan (dinamis)
input_sequence = Input(shape=(5, 3))  # 5 frame, 3 fitur per frame
x_sequence = LSTM(64)(input_sequence)

output_movement = Dense(1, activation='sigmoid', name='movement_output')(x_sequence)  # 2 kelas untuk gerakan

model_movement = Model(inputs=input_sequence, outputs=output_movement)
model_movement.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 4. Pelatihan model simbol
model_symbol.fit(X_symbol, y_symbol, epochs=5, batch_size=16)

# 5. Pelatihan model gerakan
model_movement.fit(X_movement, y_movement, epochs=5, batch_size=16)

# 6. Evaluasi Model untuk simbol
loss_symbol, accuracy_symbol = model_symbol.evaluate(X_symbol, y_symbol)
print(f"Loss (Symbol): {loss_symbol}")
print(f"Accuracy (Symbol): {accuracy_symbol}")

# 7. Evaluasi Model untuk gerakan
loss_movement, accuracy_movement = model_movement.evaluate(X_movement, y_movement)
print(f"Loss (Movement): {loss_movement}")
print(f"Accuracy (Movement): {accuracy_movement}")

# 8. Prediksi dengan model simbol
symbol_pred = model_symbol.predict(X_symbol)
print(f"Contoh prediksi simbol: {symbol_pred[:5]}")

# 9. Prediksi dengan model gerakan
movement_pred = model_movement.predict(X_movement)
print(f"Contoh prediksi gerakan: {movement_pred[:5]}")
```

```python
import cv2
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import numpy as np
import time

# Setup MediaPipe
base_options = python.BaseOptions(model_asset_path='hand_landmarker.task')
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    num_hands=1,
    running_mode=vision.RunningMode.IMAGE
)
detector = vision.HandLandmarker.create_from_options(options)

# Webcam
cap = cv2.VideoCapture(0)
prev_landmarks = None
prev_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=rgb_frame)
    result = detector.detect(mp_image)

    curr_time = time.time()
    delta_time = curr_time - prev_time
    prev_time = curr_time

    speed_text = ""

    if result.hand_landmarks:
        hand = result.hand_landmarks[0]
        h, w, _ = frame.shape

        # Gambar kerangka tangan
        mp_hands = mp.solutions.hands
        for connection in mp_hands.HAND_CONNECTIONS:
            start_idx, end_idx = connection
            x1 = int(hand[start_idx].x * w)
            y1 = int(hand[start_idx].y * h)
            x2 = int(hand[end_idx].x * w)
            y2 = int(hand[end_idx].y * h)
            cv2.line(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        for lm in hand:
            cx, cy = int(lm.x * w), int(lm.y * h)
            cv2.circle(frame, (cx, cy), 4, (255, 0, 0), -1)

        # Hitung kecepatan jika prev_landmarks ada
        if prev_landmarks:
            speed_total = 0
            for i in range(len(hand)):
                x1, y1 = hand[i].x, hand[i].y
                x0, y0 = prev_landmarks[i]
                dist = np.sqrt((x1 - x0)**2 + (y1 - y0)**2)
                speed_total += dist
            avg_speed = (speed_total / len(hand)) / delta_time
            speed_text = f"Speed: {avg_speed:.4f}"

        # Simpan landmark sekarang ke prev_landmarks
        prev_landmarks = [(lm.x, lm.y) for lm in hand]

    # Jika tidak ada tangan, jangan reset prev_landmarks, cukup skip update speed
    if avg_speed >0.5:
        print(avg_speed)
    # Tampilkan
    if speed_text:
        cv2.putText(frame, speed_text, (10, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

    cv2.imshow("Hand Speed Tracker", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

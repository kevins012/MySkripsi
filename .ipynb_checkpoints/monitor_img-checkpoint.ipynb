```python
from mediapipe import solutions
from mediapipe.framework.formats import landmark_pb2
import numpy as np
import cv2  # Pastikan OpenCV sudah diinstal

MARGIN = 10  # pixels
FONT_SIZE = 0.5
FONT_THICKNESS = 1
HANDEDNESS_TEXT_COLOR = (88, 100, 54)  # Vibrant green
LANDMARK_TEXT_COLOR = (255, 0, 0)  # Warna biru untuk teks landmark


def draw_landmarks_on_image(rgb_image, detection_result):
    hand_landmarks_list = detection_result.hand_landmarks
    handedness_list = detection_result.handedness
    annotated_image = np.copy(rgb_image)

    # Loop through the detected hands to visualize.
    for idx in range(len(hand_landmarks_list)):
      hand_landmarks = hand_landmarks_list[idx]
      handedness = handedness_list[idx]

      # Draw the hand landmarks.
      hand_landmarks_proto = landmark_pb2.NormalizedLandmarkList()
      hand_landmarks_proto.landmark.extend([
        landmark_pb2.NormalizedLandmark(x=landmark.x, y=landmark.y, z=landmark.z) for landmark in hand_landmarks
      ])
      solutions.drawing_utils.draw_landmarks(
        annotated_image,
        hand_landmarks_proto,
        solutions.hands.HAND_CONNECTIONS,
        solutions.drawing_styles.get_default_hand_landmarks_style(),
        solutions.drawing_styles.get_default_hand_connections_style())

      # Get the top left corner of the detected hand's bounding box.
      height, width, _ = annotated_image.shape
      x_coordinates = [landmark.x for landmark in hand_landmarks]
      y_coordinates = [landmark.y for landmark in hand_landmarks]
      z_coordinates = [landmark.z for landmark in hand_landmarks]
      text_x = int(min(x_coordinates) * width)
      text_y = int(min(y_coordinates) * height) - MARGIN

      # Draw handedness (left or right hand) on the image.
      cv2.putText(annotated_image, f"{handedness[0].category_name}",
                  (text_x, text_y), cv2.FONT_HERSHEY_DUPLEX,
                  FONT_SIZE, HANDEDNESS_TEXT_COLOR, FONT_THICKNESS, cv2.LINE_AA)

      # Loop setiap titik di landmark untuk menampilkan teks koordinat & indeks
      for i, (x, y, z) in enumerate(zip(x_coordinates, y_coordinates, z_coordinates)):
          text = f"x{i}"
          px, py = int(x * width), int(y * height)  # Konversi ke koordinat gambar
          cv2.putText(annotated_image, text, (px, py),
                      cv2.FONT_HERSHEY_SIMPLEX, FONT_SIZE, LANDMARK_TEXT_COLOR, FONT_THICKNESS, cv2.LINE_AA)

    return annotated_image, x_coordinates, y_coordinates, z_coordinates
```

```python
def enhance_foreground(image):
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    l = clahe.apply(l)
    lab = cv2.merge((l, a, b))
    return cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

# Fungsi untuk blur background dengan cepat
def fast_blur_background(frame):
    blurred = cv2.GaussianBlur(frame, (21, 21), 0)  # Blur seluruh gambar
    mask = np.zeros_like(frame, dtype=np.uint8)  # Mask kosong untuk tangan
    cv2.rectangle(mask, (150, 100), (490, 380), (255, 255, 255), thickness=-1)  # Area fokus tangan
    return cv2.bitwise_and(frame, mask) + cv2.bitwise_and(blurred, 255 - mask)
```

```python
import cv2
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import numpy as np

# Warna dan ukuran font untuk teks tampilan
MARGIN = 10  # pixels
FONT_SIZE = 0.5
FONT_THICKNESS = 1
TEXT_COLOR = (0, 255, 0)  # Warna hijau untuk teks koordinat
LANDMARK_TEXT_COLOR = (255, 0, 0)  # Warna biru untuk teks landmark
LINE_COLOR = (0, 255, 255)  # Warna garis (kuning-hijau)



# Buka video
cap_video = cv2.VideoCapture('test_video/paham/1.mp4')


# Pastikan video dapat dibuka
if not cap_video.isOpened():
    print("Gagal membuka video.")
    exit()

# STEP 2: Buat objek HandLandmarker
base_options = python.BaseOptions(model_asset_path='hand_landmarker.task')
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    num_hands=2,
    running_mode=vision.RunningMode.IMAGE  # Gunakan mode gambar
)
detector = vision.HandLandmarker.create_from_options(options)

print("Menampilkan video dengan deteksi tangan. Tekan 's' untuk pause, 'q' untuk keluar.")

# Variabel global untuk koordinat yang ditampilkan saat kursor diarahkan
hovered_text = ""
hovered_position = (0, 0)
paused = False  # Status video (berjalan atau pause)

# Buat jendela sebelum loop untuk menghindari error OpenCV
cv2.namedWindow("Rekaman Video dengan Hand Detection")

# Fungsi untuk menangkap event mouse
def mouse_callback(event, x, y, flags, param):
    global hovered_text, hovered_position
    if event == cv2.EVENT_MOUSEMOVE:  # Jika mouse bergerak
        hovered_text = ""
        for idx, (lx, ly, lz) in enumerate(param):
            px, py = int(lx * frame.shape[1]), int(ly * frame.shape[0])  # Skala ke ukuran gambar
            if abs(x - px) < 10 and abs(y - py) < 10:  # Jika kursor dekat titik
                hovered_text = f"ID: {idx} | X:{lx:.2f}, Y:{ly:.2f}, Z:{lz:.2f}"
                hovered_position = (px, py)
                break

# Tetapkan mouse callback ke jendela OpenCV
cv2.setMouseCallback("Rekaman Video dengan Hand Detection", mouse_callback)

while True:  # Loop utama agar video bisa diulang terus
    if not paused:
        ret, frame = cap_video.read()
        
        if not ret:  
            cap_video.set(cv2.CAP_PROP_POS_FRAMES, 0)  # Jika video habis, ulang dari awal
            continue

        # Konversi frame ke RGB untuk MediaPipe
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Konversi frame menjadi format Image MediaPipe
        mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=frame_rgb)

        # Deteksi tangan
        detection_result = detector.detect(mp_image)

        all_landmarks = []  # Simpan semua landmark untuk digunakan dalam mouse event
        if not detection_result or not detection_result.hand_landmarks:
            cv2.putText(frame, "Tidak ada tangan yang terdeteksi", 
                (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        # Gambar landmark pada frame asli
        for hand in detection_result.hand_landmarks:
            connections = mp.solutions.hands.HAND_CONNECTIONS  # Sambungan antar titik
            landmark_points = []  # Menyimpan titik untuk koneksi garis

            for idx, landmark in enumerate(hand):
                x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                all_landmarks.append((landmark.x, landmark.y, landmark.z))  # Simpan koordinat
                landmark_points.append((x, y))  # Simpan titik untuk menggambar garis
                
                # Titik landmark
                cv2.circle(frame, (x, y), 5, (0, 255, 0), -1)
                
                # Menampilkan ID setiap landmark
                cv2.putText(frame, f"{idx}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, LANDMARK_TEXT_COLOR, 1)

            # Menghubungkan titik dengan garis sesuai koneksi tangan
            for connection in connections:
                start_idx, end_idx = connection
                if start_idx < len(landmark_points) and end_idx < len(landmark_points):
                    cv2.line(frame, landmark_points[start_idx], landmark_points[end_idx], LINE_COLOR, 2)
            # Inisialisasi nilai rentang X, Y, dan Z dalam bounding box tangan
            min_x, max_x = float('inf'), float('-inf')
            min_y, max_y = float('inf'), float('-inf')
            min_z, max_z = float('inf'), float('-inf')
            
            # Loop untuk mencari rentang koordinat dalam bounding box tangan
            for landmark in hand:
                min_x, max_x = min(min_x, landmark.x), max(max_x, landmark.x)
                min_y, max_y = min(min_y, landmark.y), max(max_y, landmark.y)
                min_z, max_z = min(min_z, landmark.z), max(max_z, landmark.z)
            
            # Konversi bounding box ke skala frame OpenCV
            top_left = (int(min_x * frame.shape[1]), int(min_y * frame.shape[0]))
            bottom_right = (int(max_x * frame.shape[1]), int(max_y * frame.shape[0]))
            
            # Gambar bounding box tangan di sekitar landmark
            cv2.rectangle(frame, top_left, bottom_right, (255, 255, 0), 2)
            
            # Hitung koordinat tengah bounding box (digunakan sebagai referensi titik 0,0,0)
            center_x = (min_x + max_x) / 2
            center_y = (min_y + max_y) / 2
            center_z = (min_z + max_z) / 2
            
            # Konversi titik (0,0,0) relatif terhadap bounding box
            zero_x = int(center_x * frame.shape[1])
            zero_y = int(center_y * frame.shape[0])
            
            # Tampilkan titik (0,0,0) sebagai referensi di tengah bounding box
            cv2.circle(frame, (zero_x, zero_y), 8, (0, 0, 255), -1)
            cv2.putText(frame, "(0,0,0)", (zero_x + 10, zero_y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)
            
            # Tampilkan rentang X, Y, dan Z di layar
            cv2.putText(frame, f"X: {min_x:.2f} - {max_x:.2f}", (top_left[0], top_left[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)
            cv2.putText(frame, f"Y: {min_y:.2f} - {max_y:.2f}", (top_left[0], top_left[1] + 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)
            cv2.putText(frame, f"Z: {min_z:.2f} - {max_z:.2f}", (top_left[0], top_left[1] + 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)

    # Jika ada teks yang ditampilkan saat kursor diarahkan
    if hovered_text:
        cv2.putText(frame, hovered_text, hovered_position, cv2.FONT_HERSHEY_SIMPLEX, FONT_SIZE, TEXT_COLOR, FONT_THICKNESS)

    # Perbarui parameter yang dikirim ke callback mouse
    cv2.setMouseCallback("Rekaman Video dengan Hand Detection", mouse_callback, all_landmarks)

    # Tampilkan video dengan deteksi tangan
    cv2.imshow("Rekaman Video dengan Hand Detection", frame)

    # Tangkap input keyboard
    # Tangkap input keyboard
    key = cv2.waitKey(25) & 0xFF
    if key == ord('q'):  # Tekan 'q' untuk keluar
        break
    elif key == ord('s'):  # Tekan 's' untuk pause/unpause video
        paused = not paused
        if paused:
            print("Video dijeda. Tekan 's' untuk melanjutkan.")
        else:
            print("Melanjutkan video...")
    elif key == ord('d'):  # Tekan 'd' untuk pause tanpa mengulang video
        paused = True
        print("Video dijeda. Tekan 's' untuk melanjutkan.")


# Tutup video dan jendela
cap_video.release()
cv2.destroyAllWindows()

```

```python
import cv2
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import numpy as np

# Warna dan ukuran font untuk teks tampilan
MARGIN = 10  # pixels
FONT_SIZE = 0.5
FONT_THICKNESS = 1
TEXT_COLOR = (0, 255, 0)  # Warna hijau untuk teks koordinat
LANDMARK_TEXT_COLOR = (255, 0, 0)  # Warna biru untuk teks landmark
LINE_COLOR = (0, 255, 255)  # Warna garis (kuning-hijau)

# Nama file video yang akan diputar
video_filename = "vidio/video_22.mp4"

# Buka video
cap_video = cv2.VideoCapture(video_filename)

# Pastikan video dapat dibuka
if not cap_video.isOpened():
    print("Gagal membuka video.")
    exit()

# STEP 2: Buat objek HandLandmarker
base_options = python.BaseOptions(model_asset_path='hand_landmarker.task')
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    num_hands=2,
    running_mode=vision.RunningMode.IMAGE  # Gunakan mode gambar
)
detector = vision.HandLandmarker.create_from_options(options)

print("Menampilkan video dengan deteksi tangan. Tekan 's' untuk pause, 'q' untuk keluar.")

# Variabel global untuk koordinat yang ditampilkan saat kursor diarahkan
hovered_text = ""
hovered_position = (0, 0)
paused = False  # Status video (berjalan atau pause)

# Buat jendela sebelum loop untuk menghindari error OpenCV
cv2.namedWindow("Rekaman Video dengan Hand Detection")

# Fungsi untuk menangkap event mouse
def mouse_callback(event, x, y, flags, param):
    global hovered_text, hovered_position
    if event == cv2.EVENT_MOUSEMOVE:  # Jika mouse bergerak
        hovered_text = ""
        for idx, (lx, ly, lz) in enumerate(param):
            px, py = int(lx * frame.shape[1]), int(ly * frame.shape[0])  # Skala ke ukuran gambar
            if abs(x - px) < 10 and abs(y - py) < 10:  # Jika kursor dekat titik
                hovered_text = f"ID: {idx} | X:{lx:.2f}, Y:{ly:.2f}, Z:{lz:.2f}"
                hovered_position = (px, py)
                break

# Tetapkan mouse callback ke jendela OpenCV
cv2.setMouseCallback("Rekaman Video dengan Hand Detection", mouse_callback)

while True:  # Loop utama agar video bisa diulang terus
    if not paused:
        ret, frame = cap_video.read()
        
        if not ret:  
            cap_video.set(cv2.CAP_PROP_POS_FRAMES, 0)  # Jika video habis, ulang dari awal
            continue

        # Konversi frame ke RGB untuk MediaPipe
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Konversi frame menjadi format Image MediaPipe
        mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=frame_rgb)

        # Deteksi tangan
        detection_result = detector.detect(mp_image)

        all_landmarks = []  # Simpan semua landmark untuk digunakan dalam mouse event

        # Gambar landmark pada frame asli
        for hand in detection_result.hand_landmarks:
            connections = mp.solutions.hands.HAND_CONNECTIONS  # Sambungan antar titik
            landmark_points = []  # Menyimpan titik untuk koneksi garis

            for idx, landmark in enumerate(hand):
                x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                all_landmarks.append((landmark.x, landmark.y, landmark.z))  # Simpan koordinat
                landmark_points.append((x, y))  # Simpan titik untuk menggambar garis
                
                # Titik landmark
                cv2.circle(frame, (x, y), 5, (0, 255, 0), -1)
                
                # Menampilkan ID setiap landmark
                cv2.putText(frame, f"{idx}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, LANDMARK_TEXT_COLOR, 1)

            # Menghubungkan titik dengan garis sesuai koneksi tangan
            for connection in connections:
                start_idx, end_idx = connection
                if start_idx < len(landmark_points) and end_idx < len(landmark_points):
                    cv2.line(frame, landmark_points[start_idx], landmark_points[end_idx], LINE_COLOR, 2)
            # Inisialisasi nilai rentang X, Y, dan Z dalam bounding box tangan
            min_x, max_x = float('inf'), float('-inf')
            min_y, max_y = float('inf'), float('-inf')
            min_z, max_z = float('inf'), float('-inf')
            
            # Loop untuk mencari rentang koordinat dalam bounding box tangan
            for landmark in hand:
                min_x, max_x = min(min_x, landmark.x), max(max_x, landmark.x)
                min_y, max_y = min(min_y, landmark.y), max(max_y, landmark.y)
                min_z, max_z = min(min_z, landmark.z), max(max_z, landmark.z)
            
            # Konversi bounding box ke skala frame OpenCV
            top_left = (int(min_x * frame.shape[1]), int(min_y * frame.shape[0]))
            bottom_right = (int(max_x * frame.shape[1]), int(max_y * frame.shape[0]))
            
            # Gambar bounding box tangan di sekitar landmark
            cv2.rectangle(frame, top_left, bottom_right, (255, 255, 0), 2)
            
            # Hitung koordinat tengah bounding box (digunakan sebagai referensi titik 0,0,0)
            center_x = (min_x + max_x) / 2
            center_y = (min_y + max_y) / 2
            center_z = (min_z + max_z) / 2
            
            # Konversi titik (0,0,0) relatif terhadap bounding box
            zero_x = int(center_x * frame.shape[1])
            zero_y = int(center_y * frame.shape[0])
            
            # Tampilkan titik (0,0,0) sebagai referensi di tengah bounding box
            cv2.circle(frame, (zero_x, zero_y), 8, (0, 0, 255), -1)
            cv2.putText(frame, "(0,0,0)", (zero_x + 10, zero_y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)
            
            # Tampilkan rentang X, Y, dan Z di layar
            cv2.putText(frame, f"X: {min_x:.2f} - {max_x:.2f}", (top_left[0], top_left[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)
            cv2.putText(frame, f"Y: {min_y:.2f} - {max_y:.2f}", (top_left[0], top_left[1] + 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)
            cv2.putText(frame, f"Z: {min_z:.2f} - {max_z:.2f}", (top_left[0], top_left[1] + 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)

    # Jika ada teks yang ditampilkan saat kursor diarahkan
    if hovered_text:
        cv2.putText(frame, hovered_text, hovered_position, cv2.FONT_HERSHEY_SIMPLEX, FONT_SIZE, TEXT_COLOR, FONT_THICKNESS)

    # Perbarui parameter yang dikirim ke callback mouse
    cv2.setMouseCallback("Rekaman Video dengan Hand Detection", mouse_callback, all_landmarks)

    # Tampilkan video dengan deteksi tangan
    cv2.imshow("Rekaman Video dengan Hand Detection", frame)

    # Tangkap input keyboard
    # Tangkap input keyboard
    key = cv2.waitKey(25) & 0xFF
    if key == ord('q'):  # Tekan 'q' untuk keluar
        break
    elif key == ord('s'):  # Tekan 's' untuk pause/unpause video
        paused = not paused
        if paused:
            print("Video dijeda. Tekan 's' untuk melanjutkan.")
        else:
            print("Melanjutkan video...")
    elif key == ord('d'):  # Tekan 'd' untuk pause tanpa mengulang video
        paused = True
        print("Video dijeda. Tekan 's' untuk melanjutkan.")


# Tutup video dan jendela
cap_video.release()
cv2.destroyAllWindows()

```

```python
import os
import cv2
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

# Inisialisasi MediaPipe HandLandmarker
base_options = python.BaseOptions(model_asset_path='hand_landmarker.task')
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    num_hands=2,
    running_mode=vision.RunningMode.IMAGE
)
detector = vision.HandLandmarker.create_from_options(options)

# Variabel navigasi folder
parent_directory = 'frame_v1/'
label_list = [f for f in os.listdir(parent_directory) if os.path.isdir(os.path.join(parent_directory, f)) and f != '.ipynb_checkpoints']  # Mengabaikan folder .ipynb_checkpoints
current_label_idx = 0  # Untuk navigasi antar sub1
current_subfolder_idx = 0  # Untuk navigasi antar sub2 dalam sub1
current_image_idx = 0  # Untuk navigasi gambar dalam sub2

# Ambil daftar gambar dalam folder dan subfolder
def get_image_list(a=2):
    global current_subfolder_idx  # Ensure we're using the global variable
    sub1_folder = os.path.join(parent_directory, label_list[current_label_idx])
    sub2_folders = sorted([f for f in os.listdir(sub1_folder) if os.path.isdir(os.path.join(sub1_folder, f)) and f != '.ipynb_checkpoints'], key=lambda x: int(x))
    
    if len(sub2_folders) == 0:
        return [], ""  # Jika tidak ada subfolder, kembalikan list kosong

    # Loop untuk mencari subfolder yang valid yang mengandung gambar
    
    sub2_folder_path = os.path.join(sub1_folder, sub2_folders[current_subfolder_idx])
    image_files = sorted([f for f in os.listdir(sub2_folder_path) if f.endswith(('.jpg', '.png'))], key=lambda x: int(x.split('.')[0]))
    if image_files:  # Jika ditemukan gambar
        return image_files, sub2_folder_path
    elif a == 1:
        next_subfolder()  # Pindah ke subfolder berikutnya
    elif a ==0 :
        prev_subfolder()
    return [], ""  # Tidak ada gambar dalam subfolder yang ada

# Fungsi untuk menampilkan gambar dengan OpenCV
def show_image(image_path):
    print(f"Mencoba membuka gambar: {image_path}")  # Debug: Print path gambar yang lengkap
    image = cv2.imread(image_path)
    if image is None:
        print(f"Gagal membuka {image_path}")  # Debug: Tampilkan error jika gambar gagal dibuka
        return
    # Konversi ke RGB untuk MediaPipe
    frame_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=frame_rgb)

    # Deteksi tangan
    detection_result = detector.detect(mp_image)
    if detection_result.hand_landmarks:
        for hand in detection_result.hand_landmarks:
            # Gambar garis antara hand landmarks
            image = draw_hand_landmarks(image, hand)
    else:
        
        os.remove(image_path)
        print('Berhasil di REMOVE')
    cv2.putText(image, 
            text=image_path, 
            org=(150, 430),  # Koordinat posisi teks di gambar
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,  # Jenis font
            fontScale=1,  # Skala ukuran font
            color=(0, 255, 0),  # Warna teks (BGR format)
            thickness=2)  # Ketebalan teks

    # Tampilkan gambar dengan OpenCV
    cv2.imshow("Hand Landmarks Viewer", image)

# Fungsi untuk menggambar hand landmarks
def draw_hand_landmarks(image, hand_landmarks):
    connections = mp.solutions.hands.HAND_CONNECTIONS
    landmark_points = [(int(landmark.x * image.shape[1]), int(landmark.y * image.shape[0])) for landmark in hand_landmarks]
    for connection in connections:
        start_idx, end_idx = connection
        if start_idx < len(landmark_points) and end_idx < len(landmark_points):
            cv2.line(image, landmark_points[start_idx], landmark_points[end_idx], (0, 255, 0), 2)
    return image

# Fungsi untuk navigasi ke folder dan gambar berikutnya atau sebelumnya
def next_subfolder():
    global current_subfolder_idx
    sub1_folder = os.path.join(parent_directory, label_list[current_label_idx])
    sub2_folders = [f for f in os.listdir(sub1_folder) if os.path.isdir(os.path.join(sub1_folder, f)) and f != '.ipynb_checkpoints']
    # Gunakan prinsip wrap-around
    current_subfolder_idx = (current_subfolder_idx + 1) % len(sub2_folders)
    # Setelah berpindah ke subfolder baru, tampilkan gambar pertama dari subfolder tersebut
    show_images(1)

def prev_subfolder():
    global current_subfolder_idx
    sub1_folder = os.path.join(parent_directory, label_list[current_label_idx])
    sub2_folders = [f for f in os.listdir(sub1_folder) if os.path.isdir(os.path.join(sub1_folder, f)) and f != '.ipynb_checkpoints']
    # Gunakan prinsip wrap-around
    current_subfolder_idx = (current_subfolder_idx - 1) % len(sub2_folders)
    # Setelah berpindah ke subfolder baru, tampilkan gambar pertama dari subfolder tersebut
    show_images(0)

def next_image():
    global current_image_idx
    image_list, _ = get_image_list()
    if image_list:
        # Gunakan prinsip wrap-around
        current_image_idx = (current_image_idx + 1) % len(image_list)
    print(f"Berpindah ke gambar berikutnya: {image_list[current_image_idx]}")
    full_image_path = os.path.join(get_image_list()[1], image_list[current_image_idx])  # Path lengkap gambar
    show_image(full_image_path)  # Menampilkan gambar setelah berpindah

def prev_image():
    global current_image_idx
    image_list, _ = get_image_list(0)
    if image_list:
        # Gunakan prinsip wrap-around
        current_image_idx = (current_image_idx - 1) % len(image_list)
    print(f"Berpindah ke gambar sebelumnya: {image_list[current_image_idx]}")
    full_image_path = os.path.join(get_image_list()[1], image_list[current_image_idx])  # Path lengkap gambar
    show_image(full_image_path)  # Menampilkan gambar setelah berpindah

def next_label():
    global current_label_idx
    # Gunakan prinsip wrap-around
    current_label_idx = (current_label_idx + 1) % len(label_list)
    # Setelah berpindah ke folder sub1 baru, reset subfolder dan tampilkan gambar pertama dari sub2 pertama
    reset_subfolder()
    show_images()

def prev_label():
    global current_label_idx
    # Gunakan prinsip wrap-around
    current_label_idx = (current_label_idx - 1) % len(label_list)
    # Setelah berpindah ke folder sub1 sebelumnya, reset subfolder dan tampilkan gambar pertama dari sub2 pertama
    reset_subfolder()
    show_images()

# Fungsi untuk menampilkan gambar pertama kali
def reset_subfolder():
    global current_subfolder_idx
    # Reset subfolder index ke 0 setiap kali berpindah ke sub1 baru
    current_subfolder_idx = 0

# Fungsi untuk menampilkan gambar pertama dari folder dan subfolder
def show_images(a=2):
    image_list, subfolder_path = get_image_list(a)
    if image_list:
        global current_image_idx
        current_image_idx = 0  # Reset ke gambar pertama
        full_image_path = os.path.join(subfolder_path, image_list[current_image_idx])  # Path lengkap gambar
        print(f"Menaampilkan gambar pertama dari: {subfolder_path}")
        print(f"Path aktif sekarang: {full_image_path}")  # Menampilkan path gambar yang aktif
        show_image(full_image_path)

# Menampilkan gambar pertama dari folder dan subfolder
show_images()

# Menunggu input keyboard untuk navigasi
while True:
    key = cv2.waitKey(0) & 0xFF  # Tunggu dengan timeout 0 ms agar OpenCV tetap aktif

    if key == ord('a'):  # Pindah ke label (folder) berikutnya
        next_label()
    elif key == ord('s'):  # Pindah ke label (folder) sebelumnya
        prev_label()
    elif key == ord('m'):  # Pindah ke subfolder (sub2) berikutnya
        next_subfolder()
    elif key == ord('n'):  # Pindah ke subfolder (sub2) sebelumnya
        prev_subfolder()
    elif key == ord('t'):  # Pindah ke gambar berikutnya
        next_image()
    elif key == ord('r'):  # Pindah ke gambar sebelumnya
        prev_image()

    # Pastikan gambar ditampilkan setelah navigasi
    if key in [ord('a'), ord('s'), ord('m'), ord('n')]:  # Jika berpindah folder/subfolder
        show_images()

    if key == ord('q'):  # Keluar
        print("Menutup program...")
        break

# Tutup jendela OpenCV setelah keluar dari loop
cv2.destroyAllWindows()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Fungsi Moving Average
def moving_average(data, window_size):
    """Menghitung moving average dengan pengecekan data kosong"""
    if len(data) < window_size or len(data) == 0:
        return data  # Kembalikan data asli jika terlalu kecil atau kosong
    return np.convolve(data, np.ones(window_size) / window_size, mode='same')

# Warna tetap untuk tiap label
colors = {0: 'b', 1: 'g', 2: 'r', 3: 'c'}  # Biru, Hijau, Merah, Cyan

plt.figure(figsize=(15, 9))

data_to_corr = []  # Untuk menyimpan data yang akan dihitung korelasinya
plot_labels = []   # Label untuk data korelasi

for i in range(3):  # Loop untuk tiap label
    df_filtered = df3[df3['Label'] == i].reset_index(drop=True)  # Reset index hanya sekali di sini
    for noImg in range(1, 4):  # Loop untuk tiap 'img ke'
        data_smoothed = []  # Simpan data sementara untuk korelasi
        for a in range(1, 5):  # Loop untuk tiap kolom fitur X1, X2, X3, X4
            column_name = f'X{a}'
            if column_name in df_filtered.columns:
                data_series = df_filtered[df_filtered['img ke'] == noImg][column_name].to_numpy()
                smoothed_series = moving_average(data_series, 3)
                data_smoothed.append(smoothed_series)

        # Konversi ke array numpy dan simpan untuk korelasi
        if data_smoothed:
            data_smoothed = np.concatenate(data_smoothed)  # Gabungkan semua fitur jadi satu array
            data_to_corr.append(data_smoothed)
            plot_labels.append(f'Label {i}, Img {noImg}')
        
        # Plot garis
        plt.plot(
            range(len(data_smoothed)), data_smoothed, marker='o', linestyle='-',
            color=colors[i], label=f'Label {i}, Img {noImg}' if noImg == 1 else "_nolabel_"
        )

plt.xlabel("Feature")
plt.ylabel("Value")
plt.title("Line Chart of Series Data")
plt.grid(True)
plt.legend()
plt.show()

# === Tambahan: Korelasi antara 9 Data === #
df_corr = pd.DataFrame(data_to_corr).T.corr()  # Hitung korelasi antar data
df_corr.columns = plot_labels  # Beri nama kolom
df_corr.index = plot_labels    # Beri nama indeks



# === Visualisasi Korelasi dengan Heatmap === #
# Visualisasi Korelasi dengan Heatmap dan Nilai Angka di dalamnya
plt.figure(figsize=(10, 8))
plt.imshow(df_corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label="Correlation Coefficient")

# Menambahkan angka korelasi di dalam heatmap
for i in range(len(df_corr)):
    for j in range(len(df_corr)):
        plt.text(j, i, f"{df_corr.iloc[i, j]:.2f}", ha='center', va='center', color='black')

plt.xticks(ticks=range(len(plot_labels)), labels=plot_labels, rotation=90)
plt.yticks(ticks=range(len(plot_labels)), labels=plot_labels)
plt.title("Heatmap of Correlation between 9 Data Series")
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menghilangkan outlier dan mengganti dengan median sebelum/sesudahnya
def remove_outliers_and_replace(df, column_index, label):
    df_label = df[df['Label'] == label].copy().reset_index(drop=True)  # Filter data berdasarkan label dan reset index
    Q1 = df_label.iloc[:, column_index].quantile(0.25)  # Kuartil pertama
    Q3 = df_label.iloc[:, column_index].quantile(0.75)  # Kuartil ketiga
    IQR = Q3 - Q1  # Rentang antar-kuartil

    # Batas untuk outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifikasi nilai outlier
    outlier_indices = df_label[(df_label.iloc[:, column_index] < lower_bound) | 
                               (df_label.iloc[:, column_index] > upper_bound)].index

    # Ganti outlier dengan median dari sebelum & sesudahnya
    for idx in outlier_indices:
        if 0 < idx < len(df_label) - 1:
            # Ganti dengan median dari sebelum dan sesudahnya
            df_label.iloc[idx, column_index] = np.median([df_label.iloc[idx - 1, column_index], 
                                                          df_label.iloc[idx + 1, column_index]])
        elif idx == 0 and len(df_label) > 1:
            # Jika outlier di baris pertama, ganti dengan nilai berikutnya
            df_label.iloc[idx, column_index] = df_label.iloc[idx + 1, column_index]
        elif idx == len(df_label) - 1 and len(df_label) > 1:
            # Jika outlier di baris terakhir, ganti dengan nilai sebelumnya
            df_label.iloc[idx, column_index] = df_label.iloc[idx - 1, column_index]

    return df_label, lower_bound, upper_bound

# Warna untuk setiap kategori
colors = {1: 'red', 2: 'blue', 3: 'green'}

# Plot Data
plt.figure(figsize=(10, 6))

# Loop untuk setiap kategori
for label in [1, 2, 3]:  
    if label in df['Label'].unique():  # Pastikan label ada dalam data sebelum diproses
        df_cleaned, lower_bound, upper_bound = remove_outliers_and_replace(df, column_index=6, label=label)

        # Plot data setelah menghapus outlier
        df_cleaned.iloc[:, 6].plot(kind='line', marker='o', linestyle='-', label=f'Label {label}', color=colors[label])

        # Tambahkan garis batas outlier
        plt.axhline(y=lower_bound, color=colors[label], linestyle='--', alpha=0.7, label=f'Lower Bound {label}')
        plt.axhline(y=upper_bound, color=colors[label], linestyle='-.', alpha=0.7, label=f'Upper Bound {label}')

# Tambahkan label dan judul
plt.xlabel("Feature Index")
plt.ylabel("Value")
plt.title("Line Chart of Series Data (After Removing Outliers) with Outlier Boundaries")
plt.legend()
plt.grid(True)

# Tampilkan grafik
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the cross product
def cross_product(v1, v2):
    return np.cross(v1, v2)

# Define two vectors
a = np.array([2, 3, 4])  # Vector a
b = np.array([5, 6, 7])  # Vector b

# Calculate cross product
cross_result = cross_product(a, b)

# Plotting the vectors and their cross product
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector a, b, and the cross product
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='blue', label='Vector a', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='green', label='Vector b', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, cross_result[0], cross_result[1], cross_result[2], color='red', label='Cross product a x b', arrow_length_ratio=0.1)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add title and legend
ax.set_title("Cross Product of Vectors a and b")
ax.legend()

plt.show()
```

```python
import pandas as pd

# Data untuk 21 titik tangan (X, Y, Z untuk setiap titik dari 0 hingga 20)
data = pd.read_csv("csv/static_raw/data_try1.csv")
# Normalisasi berdasarkan titik referensi (X0, Y0, Z0)
x_ref = data['X0']
y_ref = data['Y0']
z_ref = data['Z0']

data_normalized = data.copy()

for i in range(21):
    # Normalisasi berdasarkan titik referensi (X0, Y0, Z0)
    data_normalized[f'X{i}'] = (data[f'X{i}'] - x_ref) / (x_ref.max() - x_ref.min())
    data_normalized[f'Y{i}'] = (data[f'Y{i}'] - y_ref) / (y_ref.max() - y_ref.min())
    data_normalized[f'Z{i}'] = (data[f'Z{i}'] - z_ref) / (z_ref.max() - z_ref.min())
```

```python
# Misalnya, menghitung jarak antara titik pergelangan tangan (X0, Y0, Z0) dengan ujung jari tengah (X12, Y12, Z12)
distance = np.sqrt((data['X12'] - data['X0'])**2 + (data['Y12'] - data['Y0'])**2 + (data['Z12'] - data['Z0'])**2)

# Menambahkan fitur jarak ke dalam DataFrame
data_normalized['hand_length'] = distance / distance.max()  # Normalisasi panjang tangan
```

```python
# Membuat vektor 1D untuk pergelangan tangan (X0, Y0, Z0) dan jari telunjuk (X4, Y4, Z4)
vector1 = np.array([data['X0'], data['Y0'], data['Z0']]).T  # Pergelangan tangan
vector2 = np.array([data['X4'], data['Y4'], data['Z4']]).T  # Jari telunjuk

# Menghitung cosine similarity antara dua vektor
cosine_similarity = np.einsum('ij,ij->i', vector1, vector2) / (np.linalg.norm(vector1, axis=1) * np.linalg.norm(vector2, axis=1))

# Menghitung sudut dalam radian (hasilnya dalam rentang [0, pi])
data_normalized['angle_wrist_thumb'] = np.arccos(cosine_similarity)

# Menampilkan hasil
```
